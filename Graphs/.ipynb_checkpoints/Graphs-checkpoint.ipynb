## Graph representation 

### Adjacency matrix implementation 

```python
def buildGraph1() :
    # take input for no.of nodes
    n = int(input('no.of vertices: '))
    
    # take input for no.of edges
    m = int(input('no.of edges: '))
    
    #initialize the matrix with n rows and n columns
    g = [[0 for j in range(m)] for i in range(n)] 
    
    # now take m edges as an input 
    for i in range(m) :
        u,v = map(int,input().split()) 
        g[u][v] = 1
        g[v][u] = 1
    
    # printing graph
    for i in range(m+1) :
        if i==0 :
            print(" ",end=" ")
        else :
            print(i,end=" ") 
    print()
    for i in range(n) :
        print(i,end=" ") 
        print(*g[i])
    return g
buildGraph1()
```

### Adjacency List implementation

```python
def buildGraph2() :
    # take input for no.of nodes
    n = int(input('no.of vertices: '))
    
    # take input for no.of edges
    m = int(input('no.of edges: '))
    
    #initialize the matrix with n rows and n columns
    g = [[] for i in range(n)] 
    
    # now take m edges as an input 
    for i in range(m) :
        u,v = map(int,input().split()) 
        g[u].append(v) 
        g[v].append(u)
    
    #printing graph
    for i in range(n) :
        print(i,end=" -> ") 
        print(g[i])
    return g
buildGraph2()
```

## Graph Traversals

### BFS

```python
import queue
def bfsOfGraph(n,adj) :
    vis = [0 for i in range(n)] 
    bfs = []
    q = queue.Queue()
    q.put(0) 
    vis[0] = 1
    while not q.empty() :
        node = q.get() 
        bfs.append(node)
        
        for curr in adj[node] :
            if vis[curr] == 0 :
                q.put(curr) 
                vis[curr] = 1 
    return bfs
adjlis = [[2],[2,3],[0,1,3],[1,2]]
print(bfsOfGraph(len(adjlis),adjlis))
```

### DFS

```python
def dfs(i,vis,adjlis,lis) :
    vis[i] = 1 
    lis.append(i)
    for node in adjlis[i] :
        if vis[node] == 0 :
            dfs(node,vis,adjlis,lis) 
def dfsOfGraph(n,adjlis) :
    vis = [0 for i in range(n)] 
    lis = []
    for i in range(n) :
        if vis[i] == 0 :
            dfs(i,vis,adjlis,lis)
    return lis
adjlis = [[2],[2,3],[0,1,3],[1,2]]
print(dfsOfGraph(len(adjlis),adjlis))
```
