```python
class MinHeap:
    def __init__(self,capacity):
        self.capacity = capacity
        self.size = 0
        self.array = [float('inf')]*capacity 
    def push(self,data) :
        if self.size == self.capacity :
            return 'heap is full'
        else :
            self.array[self.size] = data
            index = self.size
            #hepify up
            while True :
                parentIndex = (index-1)//2
                if parentIndex < 0 :
                    break 
                else :
                    if self.array[parentIndex] > self.array[index] :
                        #swap
                        tp = self.array[parentIndex] 
                        self.array[parentIndex] = self.array[index]
                        self.array[index] = tp
                    else :
                        break
                    index = parentIndex
            self.size += 1
    def pop(self) :
        if self.size == 0 :
            print('heap is empty') 
            return -1
        data = self.array[0] 
        self.array[0] = self.array[self.size-1] 
        index = 0
        #heapify down
        while True :
            leftChildIndex = 2*index + 1
            rightChildIndex = 2*index + 2
            if leftChildIndex >= self.capacity or rightChildIndex >= self.capacity:
                break
            smallest = index 
            if self.array[leftChildIndex] < self.array[rightChildIndex] :
                smallest = leftChildIndex
            else :
                smallest = rightChildIndex
            
            if smallest != index :
                tp = self.array[smallest]
                self.array[smallest] = self.array[index]
                self.array[index] = tp 
            else :
                break
            index = smallest
        self.array[self.size-1] = float('inf') 
        self.size -= 1
        return data
            
            
        
heap = MinHeap(10)
for i in range(10,5,-1):
    heap.push(i)
print(heap.array)
print(heap.pop())
print(heap.array)
```
